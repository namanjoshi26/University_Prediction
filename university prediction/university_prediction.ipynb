```python
x=admission_df.drop(columns=['Chance of Admit '])
y=admission_df['Chance of Admit ']
```

```python
x=np.array(x)
y=np.array(y)
```

```python
y=y.reshape(-1,1)
print(y.shape)
```

```python
#from sklearn.preprocessing import StandardScaler       # scaling the data before training the model
#scaler_x= StandardScaler()        # object for the class
#x= scaler_x.fit_transform(x)
#scaler_y= StandardScaler()        # object for the class
#y= scaler_y.fit_transform(y)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.15,random_state=11)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
# step-1: create a cross-validation scheme
folds = KFold(n_splits = 5, shuffle = True, random_state = 100)

# step-2: specify range of hyperparameters to tune
hyper_params = [{'n_features_to_select': list(range(1, 8))}]


# step-3: perform grid search
# 3.1 specify model
lm = LinearRegression()
lm.fit(x_train, y_train)
rfe = RFE(lm)             

# 3.2 call GridSearchCV()
model_cv = GridSearchCV(estimator = rfe, 
                        param_grid = hyper_params, 
                        scoring= 'r2', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)      

# fit the model
model_cv.fit(x_train, y_train)         
```

```python
# cv results
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
# plotting cv results
plt.figure(figsize=(16,6))

plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_test_score"])
plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_train_score"])
plt.xlabel('number of features')
plt.ylabel('r-squared')
plt.title("Optimal Number of Features")
plt.legend(['test score', 'train score'], loc='upper left')
```
